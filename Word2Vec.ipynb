Importing the Libraries

```python
from gensim.models import Word2Vec
import gensim
import nltk
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt_tab')
import warnings

warnings.filterwarnings(action='ignore')
```

Extracting the Data from the Souce

```python
!wget http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Sports_and_Outdoors_5.json.gz
```

```python
df = pd.read_json('reviews_Sports_and_Outdoors_5.json.gz', lines=True)
df.head()
```

Simple Preprocessing & Tokenization

```python
reviewText=df["reviewText"].apply(gensim.utils.simple_preprocess)
```

Initialize the model, Building the Vocab and Training the model

```python
model = Word2Vec(sentences=reviewText, window=10, min_count=2, workers=3)        # Initialize
model.build_vocab(reviewText)                                                    # build
model.train(reviewText, total_examples=model.corpus_count, epochs=model.epochs)  # train
```

Finding Similar Words

```python
model.wv.most_similar("bad")
```
